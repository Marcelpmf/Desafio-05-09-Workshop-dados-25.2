```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# - Gere 7 valores aleatórios entre 20°C e 35°C (temperaturas) e crie também os dias da semana.
temperatura = np.array([28, 29, 31, 32, 30, 29, 28])
dias = np.array(['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'])

data = {
    "Graus": temperatura,
    "Dias": dias
}

# - Crie um DataFrame com os dias e as temperaturas.
df = pd.DataFrame(data)
print(df)


# - Calcule a média da semana.
media = float(np.mean(temperatura))
print("\nMédia: {:.2f}°C".format(media))


# gráfico de linha mostrando a variação da temperatura ao longo da semana.
plt.plot(dias, temperatura, marker='o', label='Temperatura')
plt.axhline(media, color='r', linestyle='--', label=f'Média {media:.2f}°C')

plt.xlabel('Dias da Semana')
plt.ylabel('Temperatura (°C)')
plt.title('Variação da Temperatura ao Longo da Semana')
plt.grid(True)
plt.legend()
plt.show()

```
